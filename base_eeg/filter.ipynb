```python
import matplotlib
import pathlib
import mne
import mne_bids
matplotlib.use('Qt5Agg')
```

```python
# Read BIDS data
bids_root = pathlib.Path('raw_data/eegmatchingpennies')
bids_path = mne_bids.BIDSPath(subject='05',
                              task='matchingpennies',
                              datatype='eeg',
                              root=bids_root)

raw = mne_bids.read_raw_bids(bids_path)
```

```python
raw.plot()
```

```python
# filter
raw.load_data()
raw_filtered = raw.copy().filter(l_freq=0.3, h_freq=40)
```

```python
raw_filtered.plot()
```
